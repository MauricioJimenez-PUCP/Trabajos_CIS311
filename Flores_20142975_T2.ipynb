1. Escriba una función que transforme la temperatura de grados celsius a grados fahrenheit. La fórmula es:    
$F = 1.8*C + 32$

````python
def celsius_a_fahrenheit(temp_c):
    '''
    transforma temperatura de celsius a fahrenheit
    Input: 
        temp_c:número float
    Ouput: 
        retorna: número float
    '''
    
    temp_f = 1.8 * temp_c + 32
    
    return temp_f 
````

2. Adapte el juego de dados de la clase de control flow para que pueda aceptar cuantas simulaciones se requiera.El input será un integer del número de simulaciones que queremos. El output será un float con la probabilidad de obtener 7 sobre la probabilidad de obtener 11.

Abajo el código. 

````python
import random
````

````python
sims = 100

es_7 = 0
es_11 = 0

for i in range(sims):
    suma_dados = random.randint(1,6) + random.randint(1,6)
    if suma_dados == 7:
        es_7 +=1
    elif suma_dados == 11:
        es_11 += 1

def simulaciones_requeridas(sims):
    '''
    realiza cuantas simulaciones se requieran
    Input: 
        sims:número integer de la cantidad de simulaciones deseadas
    Ouput: 
        retorna: número float con la probabilidad de obtener 7 sobre la probabilidad de obtener 11
    '''
    
    prob7_sobre_prob11 = es_7 / es_11
    
    return prob7_sobre_prob11
````

````python
simulaciones_requeridas(1996)
````

4. Escriba una función que dé el resultado de una división de dos números, sin embargo, el resultado debe ser un número mixto. 

Los números mixtos son una representación que tiene una parte entera y una parte menor a la unidad. Por ejemplo, $50 / 7$ es representado de la siguiente manera:
$
  7\frac{1}{7}
$  

Otro ejemplo: $23 / 4$ sería 
$
  5\frac{3}{4}
$  

La función toma 2 argumentos: primero, el número a dividir, y segundo, el divisor. 

La función retorna una lista con 3 elementos: el número entero, el numerador de la fracción y el denominador de la fracción. 

Por ejemplo, ``` numero_mixto(23,4)``` da como resultado ```[5, 3, 4] ```. 

``` numero_mixto(24,4)``` da como resultado ```[6, 0, 4] ```. 

Pista: floor y módulo

````python
def numero_mixto(num, den):
    '''
    Convierte división en número mixto
    Input: 
        num: número entero
        den: número entero
    Ouput: 
        retorna: lista con 3 números
    '''
    
    lst_result = [num // den , num % den , den]
    
    return lst_result
````

````python
numero_mixto(50,7)
````

5. Escribe una función que tome como input una lista de números y los sume. En caso la suma total sea menor a 100, imprime un mensaje diciendo "Monto faltante para 100:" x . x es igual a la resta de 100 respecto a la suma total. 
En caso la suma total sea mayor a 100, el valor retornado será dicha suma total. 


````python
def suma_lista(lista_numeros):
    '''
    Suma una lista de numeros y retorna un resultado condicional:

    Input: 
        lista_numeros: lista de números
    Ouput: 
        - Si es menor a 100: nada, solo un print
        - Si es igual a 100: nada, solo un print
        - Si es mayor a 100: un número
    '''
    
    suma = sum(lista_numeros)
    
    if suma < 100:
        print("El monto faltante para 100 es", 100 - sum(lista_numeros))
    elif suma == 100:
        print("La suma es igual a 100")
    else:
        print(suma)
    
    result = sum(lista_numeros)
    
    return result
````

Ojo 2: Hay casos en los que la hora puede resultar negativa (Pista: Verlo como una resta)

````python
#Añadí dos casos más para probar la función cuando la diferencia horaria es de 12 hrs más con respecto a Lima y
#cuando es de 10 hrs menos con respecto a Lima. Hubiese elegido una GMT/UTC de -12, pero no hay ciudades en la
#Isla Howland ni en la Isla Baker.

#Dentro de la función, en la parte siguiente a
    #«convertidor_hora = suma» puse
    #if suma < 0:
        #convertidor_hora = 24 + suma
    #Usé un «24 + suma» en vez de un «24 - suma» porque un «suma» negativo se iba a convertir en + por el «24 - (-suma)» 

def convertidor_hora(hora_Lima, ciudad_extranjera):
    '''
    Convierte la hora de Lima a la hora de cuatro ciudades de países diferentes

    Input: 
        hora_Lima: número integer
        ciudad_extranjera: string
    Ouput: 
        hora_ciudad_extranjera: número integer
    '''
    GMT_ciudad_extranjera = 0
    
    if ciudad_extranjera == "PortVila_Vanuatu":
        GMT_ciudad_extranjera = +11
    elif ciudad_extranjera == "Roseau_Dominica":
        GMT_ciudad_extranjera = -4
    elif ciudad_extranjera == "Abuya_Nigeria":
        GMT_ciudad_extranjera = +1
    elif ciudad_extranjera == "Pekín_China":
        GMT_ciudad_extranjera = +8
    elif ciudad_extranjera == "Auckland_NuevaZelanda":
        GMT_ciudad_extranjera = +12
    elif ciudad_extranjera == "Honolulu_Hawai_EstadosUnidos":
        GMT_ciudad_extranjera = -10
    
    suma = (hora_Lima + 5) + (GMT_ciudad_extranjera)
    
    convertidor_hora = suma
    
    if suma < 0:
        convertidor_hora = 24 + suma
    elif suma > 24:
        convertidor_hora = suma - 24

    if convertidor_hora == 0:
        print("En", ciudad_extranjera, "son las", convertidor_hora, "hrs")
    elif convertidor_hora == 1:
        print("En", ciudad_extranjera, "es la", convertidor_hora, "am")
    elif (convertidor_hora > 1) and (convertidor_hora < 12):
        print("En", ciudad_extranjera, "son las", convertidor_hora, "am")
    elif convertidor_hora == 12:
        print("En", ciudad_extranjera, "son las", convertidor_hora, "m")
    elif convertidor_hora > 12:
        print("En", ciudad_extranjera, "son las", convertidor_hora, "hrs")
    
    return convertidor_hora
````
